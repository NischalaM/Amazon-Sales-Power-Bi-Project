```python
# Assuming df is your DataFrame with 'Order Date' and 'Ship Date' as datetime columns

# Convert datetime columns to matplotlib dates
df['Order Date'] = pd.to_datetime(df['Order Date'])  # Ensure dates are in datetime format
df['Order Date_num'] = matplotlib.dates.date2num(df['Order Date'])

df['Ship Date'] = pd.to_datetime(df['Ship Date'])  # Ensure dates are in datetime format
df['Ship Date_num'] = matplotlib.dates.date2num(df['Ship Date'])

# Now plot using the numerical date representations
df.plot(x='Order Date_num', y='Ship Date_num')
plt.show()
```
